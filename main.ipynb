```python
from ultralytics import YOLO

model = YOLO("yolov9e.pt")

model.train(
    data="X:/Epics/Acidental/data.yaml",  
    epochs=25,                          
    imgsz=250,                            
    batch=6,
    workers=8,
    device=0,                             
)
```

```python
from ultralytics import YOLO

model2 = YOLO("runs/detect/train1/weights/best.pt")
```
